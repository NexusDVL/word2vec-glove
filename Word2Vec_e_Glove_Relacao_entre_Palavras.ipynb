# ***Word2Vec:***

```python
import gensim.downloader as api
```

```python
w2v_model = api.load("word2vec-google-news-300")
```

```python
word_vector = w2v_model['car']
print("Vetor Word2Vec para a palavra 'car':", word_vector)
```

```python
similar_words = w2v_model.most_similar('car', topn=5)
print("Palavras mais similares a 'car':", similar_words)
```

# ***Glove:***

```python
import gensim.downloader as api
```

```python
glove_model = api.load("glove-wiki-gigaword-100")
```

```python
word_vector = glove_model['car']
print("Vetor GloVe para a palavra 'car':", word_vector)
```

```python
similar_words = glove_model.most_similar('car', topn=5)
print("Palavras mais similares a 'car':", similar_words)
```


**Conclusão:** Os resultados foram bem parecidos, como podemos perceber logo abaixo:


Word2Vec: vehicle, cars, truck, SUV, minivan.

Glove: vehicle, cars, truck, driver, driving.
